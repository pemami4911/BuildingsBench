```python
import glob
import pyarrow.parquet as pq

dataset_path = Path(os.environ.get('BUILDINGS_BENCH', ''))
puma_files = list((dataset_path / 'Buildings-900K-test' / '2021').glob('*comstock*/*/*/*/*/*.parquet'))

test_ids = set()
for pum_file in puma_files:
    print(pum_file)
    bldg_ids = pq.read_table(pum_file).to_pandas().columns[1:]

    for bldg_id in bldg_ids:
        test_ids.add(int(bldg_id))
```
